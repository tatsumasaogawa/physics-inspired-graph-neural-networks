# データの加工

## 1-layer gnn

```python
for n in range(100, 1001, 100):
    for d in range(3, 6, 2):
        mc_n_d = []
        time_n_d = []
        for seed_value in range(20):
            input_data = open('../1_layer_gnn/data/1gnn_n' + str(n) + '_d' + str(d) + '_seed' + str(seed_value) + '.dat', 'r')

            # 一行ずつ読み込んでは表示する
            for rows in input_data:
                # コメントアウト部分を省く処理
                if rows[0]=='#':
                    continue
                
                # 値を変数に格納する
                row = rows.rstrip('\n').split(' ')
                mc_size = float(row[0])
                tot_time = float(row[2])

                mc_n_d.append(mc_size)
                time_n_d.append(tot_time)

            # ファイルを閉じる
            input_data.close()
        
        ave_mc = np.mean(mc_n_d)
        std_mc = np.std(mc_n_d)
        ave_time = np.mean(time_n_d)
        std_time = np.std(time_n_d)

        outfile = open('../1_layer_gnn/processed_data/1gnn_n' + str(n) + '_d' + str(d) + '.dat','w')
        outfile.write('# 01: average mc size 02: std of mc size 03: average solution time 04: std of solution time\n')
        outfile.write(str(ave_mc) + ' ' + str(std_mc) + ' ' + str(ave_time) + ' ' + str(std_time) + '\n')
        outfile.close()

for n in range(2000, 10001, 1000):
    for d in range(3, 6, 2):
        mc_n_d = []
        time_n_d = []
        for seed_value in range(20):
            input_data = open('../1_layer_gnn/data/1gnn_n' + str(n) + '_d' + str(d) + '_seed' + str(seed_value) + '.dat', 'r')

            # 一行ずつ読み込んでは表示する
            for rows in input_data:
                # コメントアウト部分を省く処理
                if rows[0]=='#':
                    continue
                
                # 値を変数に格納する
                row = rows.rstrip('\n').split(' ')
                mc_size = float(row[0])
                tot_time = float(row[2])

                mc_n_d.append(mc_size)
                time_n_d.append(tot_time)

            # ファイルを閉じる
            input_data.close()
        
        ave_mc = np.mean(mc_n_d)
        std_mc = np.std(mc_n_d)
        ave_time = np.mean(time_n_d)
        std_time = np.std(time_n_d)

        outfile = open('../1_layer_gnn/processed_data/1gnn_n' + str(n) + '_d' + str(d) + '.dat','w')
        outfile.write('# 01: average mc size 02: std of mc size 03: average solution time 04: std of solution time\n')
        outfile.write(str(ave_mc) + ' ' + str(std_mc) + ' ' + str(ave_time) + ' ' + str(std_time) + '\n')
        outfile.close()
```

# プロット作成

```python
def ub(d, n):
    return (d / 4 + 0.7632 * np.sqrt(d / 4)) * n
```

## 最大カットサイズとその上界に対する比

```python
from matplotlib import pyplot as plt

x_1 = np.array(list(range(100, 1001, 100)) + list(range(2000, 10001, 1000)))
x_2 = np.array(list(range(100, 1001, 100)) + list(range(2000, 2001, 1000)))
x_3 = np.array(list(range(100, 1001, 100)))

fig, ax = plt.subplots()
ax.errorbar(x_1, ave_mc_1gnn[0], yerr=std_mc_1gnn[0], capsize=3)
ax.errorbar(x_1, ave_mc_2gnn[0], yerr=std_mc_2gnn[0], capsize=3)
ax.errorbar(x_2, ave_mc_mfa[0], yerr=std_mc_mfa[0], capsize=3)
ax.plot(x_1, ub(3, x_1))
ax.set_xlabel("n")
ax.set_ylabel('MC size')
plt.legend(["upper bound", "1-layer GNN", "2-layer GNN", "MFA"])
plt.show()
```

```python
def func(X, a, b):
    Y = a * X ** b
    return Y
```
